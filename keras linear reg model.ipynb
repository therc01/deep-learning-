```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
cl = concrete_data.columns
predictor = concrete_data[cl[cl != 'Strength']]
target = concrete_data['Strength']
print(predictor.shape)
print(target.shape)
```

```python
# Part 2 starts here: normalised inputs
predictors_norm = (predictor - predictor.mean()) / predictor.std()
predictors_norm.head()
```

```python
model = func()
error_mean_norm = []
error_std_norm = []
for i in range(50):
    X_train,  X_test,Y_train, Y_test = train_test_split(predictors_norm,target,test_size = 0.3)
       
    model.fit(X_train, Y_train, epochs = 50, verbose = 2)
    predictions = model.predict(X_test)
    error = mean_squared_error(Y_test, predictions)
    
    error_mean_norm.append(error.mean())
    error_std_norm.append(error.std())
    print(f'epoch {i}')
    
```

```python
error_mean_norm_ = np.sum(error_mean_norm)/50
error_mean_norm_          # mean of normalised input
```

```python
# Part 4 using more layered network
```
